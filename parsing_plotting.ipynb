```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np


csv_file_directory="./csv_files/cloudlab_200s_emulation_off/"
file_names = os.listdir(csv_file_directory)
GD_files=[i for i in file_names if "GD_200_" in i]
GD_fast_files=[i for i in file_names if "GD_FAST_" in i]
BO_files=[i for i in file_names if "BO_" in i]

GD_thr=[]
GD_plr=[]

GD_fast_thr=[]
GD_fast_plr=[]

BO_thr=[]
BO_plr=[]

for file in GD_files:
    data = pd.read_csv(csv_file_directory+file,sep='\t')
    df = pd.DataFrame(data)
    GD_thr.append(df["curr_thrpt"].mean())
    GD_plr.append(df["packet_loss_rate"].sum())

    
for file in GD_fast_files:
    data = pd.read_csv(csv_file_directory+file,sep='\t')
    df = pd.DataFrame(data)
    GD_fast_thr.append(df["curr_thrpt"].mean())
    GD_fast_plr.append(df["packet_loss_rate"].sum())

    
for file in BO_files:
    data = pd.read_csv(csv_file_directory+file,sep='\t')
    df = pd.DataFrame(data)
    BO_thr.append(df["curr_thrpt"].mean())
    BO_plr.append(df["packet_loss_rate"].sum())
```

```python
print(GD_thr)
print(GD_plr)
print(GD_fast_thr)
print(GD_fast_plr)
print(BO_thr)
print(BO_plr)
```

```python
throughput=np.asarray([GD_thr,GD_fast_thr,BO_thr])
print(np.transpose(throughput))
print(throughput,throughput.shape)
print(throughput.reshape(-1,3))
```

```python
from matplotlib import pyplot as plt
import numpy as np
plt.rcParams["font.serif"] = "Times"
plt.rcParams["font.family"] = "serif"


throughput=np.asarray([GD_thr,GD_fast_thr,BO_thr])
plr=np.asarray([GD_plr,GD_fast_plr,BO_plr])


fig, ax1 = plt.subplots(figsize=(10, 5))

x_axis = [
    "GD","BO","RL(PP0)"
]


# ax1.set_xlabel('algorithm')
# ax1.set_ylabel('throughput')
res1 = ax1.boxplot(
    np.transpose(throughput),positions =np.arange(3)-0.25, widths=0.4,
    patch_artist=True,
)
for element in ['boxes', 'whiskers', 'fliers', 'means', 'medians', 'caps']:
    plt.setp(res1[element], color='k')

for patch in res1['boxes']:
    patch.set_facecolor('tab:blue')



ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
# ax2.set_ylabel('Aggregated PLR')
res2 = ax2.boxplot(
    np.transpose(plr), positions =np.arange(3)+0.25,widths=0.4,
    patch_artist=True,
)
#from https://stackoverflow.com/a/41997865/2454357
for element in ['boxes', 'whiskers', 'fliers', 'means', 'medians', 'caps']:
    plt.setp(res2[element], color='k')

for patch in res2['boxes']:
    patch.set_facecolor('tab:orange')

ax1.set_ylim([500, 1000])
ax2.set_ylim([0, 1])
ax1.set_xticks(np.arange(3))
ax1.set_xticklabels(x_axis)
ax1.xaxis.set_tick_params(labelsize=30)
ax1.yaxis.set_tick_params(labelsize=30)
ax2.yaxis.set_tick_params(labelsize=30)
ax1.set_ylabel('Throughput in \nMbps', fontweight ='bold',fontsize=30)
ax2.set_ylabel('Aggregated \nPLR', fontweight ='bold',fontsize=30)
ax1.set_xlabel('Algorithms', fontweight ='bold',fontsize=30)
ax1.legend([res1["boxes"][0], res2["boxes"][0]], ['Throughput', 'PLR'], loc='center right',fontsize=20)
fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
# fig.savefig('test.eps', format='eps',bbox_inches='tight',dpi=1000)
```
